```python
# Dependencies
import pandas as pd
import plotly.express as plt
import os
import statsmodels.api as sm
```

```python
bikecounts_file = os.path.join('../','Resources', 'bikecounts.csv')
bikecounts_df = pd.read_csv(bikecounts_file)
bikecounts_df['Day'] = pd.to_datetime(bikecounts_df['Day'])
bikecounts_df.head()
```

```python
#hypothesis: Low temps will cause less bike traffic on the bridge

bikecounts_df.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
bikecounts_df['Median Temp (°F)']=(bikecounts_df['High Temp (°F)'] + bikecounts_df['Low Temp (°F)'])/2
```

```python
def remove_unwanted(value):
    for unwanted_charater in ['T', '(S)', ' ']:
        value=value.replace(unwanted_charater, '')
    return value.strip()
```

```python
bikecounts_df['Precipitation']= bikecounts_df['Precipitation'].apply(remove_unwanted)
bikecounts_df['Precipitation']= bikecounts_df['Precipitation'].replace('',0).astype('float64')
bikecounts_df.head()
```

```python
#Null hypothesis: Low temps and precipitation doesnt effect the number of bikers on the bridge
```

```python
independent_variables=bikecounts_df[['Median Temp (°F)', 'Precipitation']]
dependent_value=bikecounts_df['Brooklyn Bridge']
X=sm.add_constant(independent_variables)
regression_model=sm.OLS(dependent_value,X)
results=regression_model.fit()
```

```python
print(results.summary())
```

```python
results.summary2()
```

```python
#Solution: The p-vaule is close to 0 so the null hypothesis is wrong. Temp and precipitation effects the amount of bikers on the bridge.
```
